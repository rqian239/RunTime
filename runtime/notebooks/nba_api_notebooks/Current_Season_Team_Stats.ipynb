```python


# Import statements
import pandas as pd
from datetime import datetime
from nba_api.stats.static import teams

from nba_api.stats.endpoints import teamgamelogs
from nba_api.stats.endpoints import teaminfocommon
```

```python

teamGameLogs = teamgamelogs.TeamGameLogs(season_nullable='2023-24')
df = teamGameLogs.get_data_frames()[0]

df
```

```python
## now time to get the current average stats of the last 10 games using my code from last script

def average_stats_last_10_games(group):
    last10 = group.tail(10)
    averages = last10.mean(numeric_only=True)
    return averages
```

```python
avg_stats_by_team = df.groupby('TEAM_NAME').apply(average_stats_last_10_games)
avg_stats_by_team
```
