```python
import pandas as pd
```

```python
df = pd.read_csv("nba_games_runtime.csv", index_col=0)
df
```

```python
# Remove extraneous columns
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]

phrase = "_opp"
remove = [l for l in df.columns if phrase in l]
df_cleaned = df.drop(columns=remove)
df_cleaned.head()
df = df_cleaned
df
```

```python
df = df.sort_values(by=['team', 'date'])
df
```

```python
def average_stats_last_10_games(group):
    last10 = group.tail(10)
    averages = last10.mean(numeric_only=True)
    return averages
```

```python
avg_stats_by_team = df.groupby('team').apply(average_stats_last_10_games)
avg_stats_by_team
```
