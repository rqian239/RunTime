```python
# Import statements
import pandas as pd
from datetime import datetime
from nba_api.stats.static import teams

from nba_api.stats.endpoints import TeamDetails
from data import teams
from nba_api.stats.library import data
```

```python
all_teams = teams.get_teams()

# Create lists to store team data
team_data = []

# Iterate over each team
for team in all_teams:
    team_id = team['id']
    abbreviation = team['abbreviation']
    nickname = team['nickname']
    year_founded = team['year_founded']
    city = team['city']
    full_name = team['full_name']
    state = team['state']
    
    # Check if the key 'championship_years' exists
    if 'championship_years' in team:
        championship_years = team['championship_years']
    else:
        championship_years = []
    
    # Append team data to the list
    team_data.append([team_id, abbreviation, nickname, year_founded, city, full_name, state, championship_years])

# Create DataFrame from the collected team data
df = pd.DataFrame(team_data, columns=['TEAM_ID', 'ABBREVIATION', 'NICKNAME', 'YEARFOUNDED', 'CITY', 'FULL_NAME', 'STATE', 'CHAMPIONSHIP_YEARS'])

# Print the DataFrame
print(df)
```
